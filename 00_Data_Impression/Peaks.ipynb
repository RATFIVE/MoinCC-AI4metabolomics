```python
from classes import DataParser
from classes import PeakFinder
import numpy as np
import pandas as pd

# import gaussian noise filter
from scipy.ndimage import gaussian_filter
```

```python
dataparser = DataParser()
data_list = dataparser.load_data()

for i in range(len(data_list)):
    df = dataparser.load_df(data_list[i])
    

    print(f"Data {i}: {data_list[i]}")
    # sum all columns from 1 to len(df.columns)
    df['sum'] = df.iloc[:, 1:].sum(axis=1)

    x = df.iloc[:, 0].values
    y = df['sum'].values

    
    

    peakfinder = PeakFinder()
    peaks, prob, tresh = peakfinder.peaks_finder(y)
    


    peak_pos_list = np.array([2.323, 6.653, 9.094, 8.876, 8.420, 7.772, 2.468, 4.7, 1.201, 4.368, 2.475, 9.031, 8.714, 8.376, 7.659, 1.2261, 1.9775])
    
    true_peaks = []
    # check if peak is close to the peak_pos_list
    bound = 0.2
    for peak in peaks:
        for peak_pos in peak_pos_list:
            if x[peak] - bound <= peak_pos <= x[peak] + bound:
                #print(f"Peak {peak} is close to peak_pos {peak_pos}")
                true_peaks.append(peak)


    import matplotlib.pyplot as plt
    plt.plot(x, y, label='Data', color='blue')
    # plot horizontal line at threshold
    #plt.axhline(tresh, color='green', linestyle='--')

    # plot vertical lines at peaks
    for peak in peak_pos_list:
        plt.axvline(peak, color='orange', linestyle='--', alpha=0.8, label='Metabolite Position')

    for peak in true_peaks:
        plt.axvline(x[peak], color='green', linestyle='--', alpha=0.8, label='Peak in Range of Metabolite Position')
    
    plt.scatter(x[true_peaks], y[true_peaks], color='red', label='Found Peaks')
    plt.title(f"Data {i}")
    # Legend with grouped labels
    handles, labels = plt.gca().get_legend_handles_labels()
    by_label = dict(zip(labels, handles))
    plt.legend(by_label.values(), by_label.keys())
    plt.show()
```
