```python
import pandas as pd
import numpy as np

import seaborn as sns
import matplotlib.pyplot as plt

from plotly.subplots import make_subplots
```

```python
#load data
data = pd.read_csv("../Data/FA _20240215_2H_Yeast_Pyruvate-d3_5.csv")
data
```

```python
data.shape
```

```python
#create first three spectra vizualizations
data.iloc[:, 1::20]
```

```python
# get impression of data; visualize every 20th spectra 
plt.figure(figsize=(10, 6))

for i in range(1,len(data.columns), 20):
    sns.lineplot(x=data.iloc[:, 0], y=data.iloc[:, i], label=f'{i}')

plt.xlabel( 'chemical shift')
plt.ylabel('Intensity')
plt.legend(title='shown columns')
plt.show()
```

```python
# sum up all spectra
sum_spec = pd.DataFrame()
sum_spec['chem_shift'] = data['Unnamed: 0']
sum_spec['Intensity_sum']= data.iloc[:, 1:].sum(axis=1)
```

```python
#visualize sum of spectra
plt.figure(figsize=(10, 6))
sns.lineplot(sum_spec, x= 'chem_shift', y= 'Intensity_sum')
```

Peaks are very clearly visible, noise is reduced. Overlapping peaks, such as the first one are still recognizable.

--> can be used for initial fining if peak positions
--> maybe get derivative of graphs

```python
# calculate derivatives
sum_spec['1st derivative'] = np.gradient(sum_spec['Intensity_sum'], sum_spec['chem_shift'])
sum_spec['2nd derivative'] = np.gradient(sum_spec['1st derivative'], sum_spec['chem_shift'])
```

```python
plt.figure(figsize=(10, 6))

sns.lineplot(sum_spec, x= 'chem_shift', y= '1st derivative', label= '1st derivative')
sns.lineplot(sum_spec, x= 'chem_shift', y= '2nd derivative', label= '2nd derivative')

plt.axhline(0, color="black", linewidth=0.5)

plt.show()
```

```python
fig, (ax1, ax2) = plt.subplots(2, 1, sharex=True, figsize=(10, 8))

sns.lineplot(sum_spec, x= 'chem_shift', y= 'Intensity_sum',ax= ax1)

plt.ylim([-0.4e9, 0.4e9])
sns.lineplot(sum_spec, x= 'chem_shift', y= '1st derivative', ax= ax2, label= '1st derivative')
sns.lineplot(sum_spec, x= 'chem_shift', y= '2nd derivative', ax= ax2,label= '2nd derivative')

plt.axhline(0, color="black",   linewidth=0.5)

ax1.set_title('Sum of all spectra')
ax2.set_title('Derivatives')

plt.xlabel('x')

plt.tight_layout()
plt.show()
```
